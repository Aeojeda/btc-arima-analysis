# Capítulo 10: Conclusiones Generales

## 10.1 Sobre la serie BTC-USD

La serie histórica de Bitcoin (2014–presente) es un proceso **integrado de orden 1 — I(1)**, confirmado por la convergencia de los tests ADF y KPSS:

- **ADF en nivel:** estadístico = −1.15, p = 0.69 → no rechaza raíz unitaria
- **KPSS en nivel:** estadístico = 7.50 >> valor crítico al 1% (0.739) → rechaza estacionariedad
- **Primera diferencia:** ambos tests confirman estacionariedad → $d = 1$

Los log-retornos exhiben **clusters de volatilidad** (efecto ARCH), estructura que ARIMA no captura. Un modelo ARIMA-GARCH sería más apropiado para modelar completamente la distribución condicional.

## 10.2 Sobre la selección del modelo

| Criterio | Modelo predominante | Penalización por parámetro ($n \approx 3800$) |
|----------|--------------------|-----------------------------------------|
| AIC  | ARIMA(1,1,0) / ARIMA(2,1,1) | $2k = 2$ |
| BIC  | ARIMA(1,1,0) | $k\ln(3800) \approx 8.24k$ |
| HQIC | ARIMA(1,1,0) | $2k\ln(\ln(3800)) \approx 4.17k$ |

Los tres criterios convergen al mismo modelo de orden bajo para la mayoría de horizontes, reflejando que la dinámica de BTC diferenciado tiene **memoria de corto alcance**.

## 10.4 Sobre los residuales

- **Independencia (Ljung-Box):** En la mayoría de los modelos se acepta H₀ — los residuales son ruido blanco. Cuando se rechaza, indica que el modelo no capturó toda la estructura temporal.
- **Normalidad (JB, SW):** Típicamente rechazada — los residuales de BTC presentan **leptocurtosis** (colas pesadas). Esto invalida los intervalos de confianza basados en distribución normal, pero no afecta la validez de las predicciones puntuales.

## 10.6 Limitaciones y extensiones

:::{admonition} Limitaciones del análisis ARIMA para BTC
:class: warning

1. **Heterocedasticidad condicional:** ARIMA asume varianza constante. Los clusters de volatilidad de BTC requieren modelos GARCH para capturar la dinámica de la varianza condicional.

2. **Evaluación con ventana única:** Los resultados son sensibles al período específico de test. Una evaluación más robusta requeriría **walk-forward validation** con múltiples ventanas.

3. **Linealidad:** ARIMA es un modelo lineal. Las no-linealidades en BTC (asimetría de respuesta a shocks positivos/negativos) no son capturadas.

4. **Hipótesis de mercado eficiente:** El MAPE ~2% del rolling no implica que el modelo sea rentable en trading — los costos de transacción y el spread suelen superar ese margen.
:::

:::{admonition} Extensiones posibles
:class: note

- **ARIMA-GARCH:** Modela simultáneamente la media y la varianza condicional
- **SARIMA:** Si se detecta estacionalidad semanal en la serie diferenciada
- **VAR:** Incorporar variables exógenas (volumen, dominancia BTC, tasas de interés)
- **LSTM / Transformer:** Modelos de deep learning para capturar dependencias no lineales de largo plazo
:::
